# Matplotlib workflow

```python
#import matplotlib and get it ready to plot in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#prepare data
x = [1,2,3,4]
y = [22,12,5,20]

#setup plot
fig,ax = plt.subplots(figsize=(10,5))

#plot data
ax.plot(x,y);

#customize plot 
ax.set(title = 'Simple plot',
      xlabel="X axis label",
      ylabel="Y axis label")

#save the graph to your computer
#fig.savefig('images/sampleplot.png')
```

```python
#make a plot from a dictionary
nut_prices = {'Almond butter': 10,
             'Peanut butter': 7,
             'Cashew butter': 11}
fig, ax = plt.subplots()
ax.bar(nut_prices.keys(), nut_prices.values());
ax.set(title= 'Butter prices',
      ylabel='price($)');
```

```python
#bar chart from dictionary
agenadsex = {'females':20,
            'males':25}
fig, ax = plt.subplots()
ax.bar(agenadsex.keys(),agenadsex.values())
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.scatter(lnspace, lnspace//2);
```

```python
#horizontal bar graph
fig, ax = plt.subplots()
ax.barh(list(nut_prices.keys()), list(nut_prices.values()));
#convert both axis' to a list or therewould be an error
```

### Subplots and the two methods

```python
#make 4 different axes on one figure and plot on all of them
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(12,9))
ax1.pie(agenadsex.values(), labels=agenadsex.keys())
ax1.set(title='pie chart')
ax2.bar(nut_prices.keys(), nut_prices.values())
ax2.set(ylabel='price($)',
       title='bar chart');
ax3.plot([1,2,4,6,45,50],[21,11,4,6,15,21])
ax3.set(ylabel='y axis',
       xlabel='x axis',
       title='line graph');
ax4.scatter([1,2,4,6,45,50],[21,11,4,6,15,21])
ax4.set(ylabel='yaxis',
       xlabel='xaxis',
       title='Scatter plot');
#fig.savefig('images/first_subplot')
```

```python
#second subplot method
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
ax[0,0].plot([1,2,4,6,45,50],[21,11,4,6,15,21])
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_prices.keys(), nut_prices.values())
ax[1,1].hist(np.random.randn(100));
#this second method uses list indexing to access the different axes
```

#### Using another dataset to plot

```python
heart_disease = pd.read_csv("data/13.1 heart-disease.csv")
heart_disease.head()
```

```python
heart_disease['age'].plot.hist();
```

### the OO method of plotting 

```python
heart_disease.head()
```

```python
#get patients over 50
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
#pyplot method of plotting
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
#ax.set_xlim(45,100)   #x axis number limit
```

```python
# OO method afresh
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
#customize the plot
ax.set(title='Heart disease and Cholesterol Levels',
      xlabel='Age',
      ylabel='Cholesterol')
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                              figsize=(15,10))

ax1.scatter(x=over_50['age'],
                   y=over_50['thalach'],
                   c=over_50['target'])
ax1.set(title='Heart disease and heartbeat rate',
      xlabel='Age',
      ylabel='Heartbeat rate')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(over_50['thalach'].mean(),
          linestyle='--')
myplot2 = ax2.hist(over_50['trestbps'])
ax2.set(ylabel='Frequency', xlabel='Thalach');
#fig.savefig('images/myplot_practice.png')
```

```python
over_50.head(10)
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True  #this line makes the whole figure to share just one x axis 
                             )
#add data to axis 1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
#customize ax1
ax1.set(title='Heart disease and cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol')
#add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
#add a horizontal line at the mean
ax1.axhline(y=over_50['chol'].mean(),
           linestyle='--')

#add data to ax2
thal = ax2.scatter(x=over_50['age'],
                  y=over_50['thalach'],
                  c=over_50['target'])
#customize ax2
ax2.set(title='Heart disease and maximum heartrate',
       xlabel="Age",
       ylabel='Maximum heartrate')
#add legend to ax2
ax2.legend(*scatter.legend_elements(), title='Target')
#add horizontal line at the mean
ax2.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

#add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```

```python
#change the plot style
plt.style.use('seaborn-whitegrid')

#object method of plotting below
fig, (ax1,ax2) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True  #this line makes the whole figure to share just one x axis 
                             )
#add data to axis 1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')   #cmap changes the color scheme
#customize ax1
ax1.set(title='Heart disease and cholesterol levels',
      xlabel='Age',
      ylabel='Cholesterol')
#change the x axis limit on ax1
ax1.set_xlim([50,80])

#add legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
#add a horizontal line at the mean
ax1.axhline(y=over_50['chol'].mean(),
           linestyle='--')

#add data to ax2
thal = ax2.scatter(x=over_50['age'],
                  y=over_50['thalach'],
                  c=over_50['target'],
                  cmap='winter')
#customize ax2
ax2.set(title='Heart disease and maximum heartrate',
       xlabel="Age",
       ylabel='Maximum heartrate')
#change the x and y axis limit on ax2
ax2.set_xlim([50,80])
ax2.set_ylim([60,200])
#add legend to ax2
ax2.legend(*scatter.legend_elements(), title='Target')
#add horizontal line at the mean
ax2.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

#add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
#save the figure
#fig.savefig('images/heart_desease_analysis.png')
```
